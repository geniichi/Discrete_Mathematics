```python
from matplotlib_venn import venn2
```

```python
def drawVenn2(A,B,labels = ('A', 'B')):
    '''
    listing elements in Venn diagram of two sets
    '''
    v2=venn2([A,B],set_labels = labels)
    v2.get_label_by_id('01').set_text(', '.join(map(str,B-A)))
    v2.get_label_by_id('10').set_text(', '.join(map(str,A-B)))
    if A&B:
        v2.get_label_by_id('11').set_text(', '.join(map(str,A&B)))
```

# Providing Identities

#### Given: $x \in A \land x \in B$ 

1. $A \setminus B = A \cap B^{c}$

A = {'emu otori', 'hoshino takanashi'} \
B = {'emu otori','yui hirasawa'} 

- so if $ x \in A $ is: $A$={'emu_otori', 'hoshino_takanashi'}
- and $x \in B$ is: $B$={'emu_otori','yui_hirasawa'}
- then $A \setminus B$ is: $x \in$ {hoshino_takanashi}

So the answer to the left half is: {hoshino_takanashi}

- so if $B^{c}$ is: $B^{c}$={'hoshino_takanashi'}
- and $x \in A$ is: $A$={'emu_otori', 'hoshino_takanashi'}
- then $A \cap B^{c}$ is: {hoshino_takanashi}

So the answer to the right half is: {hoshino_takanashi}

### Therefore this statement; $A \setminus B = A \cap B^{c}$ is $\boxed{correct}$
since both halves are: {hoshino_takanashi} = {hoshino_takanashi}

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori','yui\n hirasawa'} 

drawVenn2(A,B,labels=('A', 'B'))

# A \ B 
difference = A.difference(B)

# Bᶜ
complement_B = set()
for element in A:
    if element not in B:
        complement_B.add(element)

# A ∩ Bᶜ
intersection = A.intersection(complement_B)

# difference - intersection
is_correct = difference == intersection
print("Is A \setminus B = A \cap B^{c} correct?", is_correct)
```

2. B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$)

A = {'emu_otori', 'hoshino_takanashi'} \
B = {'emu_otori','yui_hirasawa'} 

- so if $ x \in B $ are: $B$={'emu_otori','yui_hirasawa'}
- and $x \in A$ are: $A$={'emu_otori', 'hoshino_takanashi'}
- then $A \cap B$ is: $x \in$ {emu_otori}

So the answer to the left half is: {emu_otori}

- if $B$ is: $B$={'emu_otori','yui_hirasawa'}
- and $x \in A^{c}$ is: $A^{c}$={'yui_hirasawa'}
- then $B \cap A^{c}$ is: {'yui_hirasawa'}

So the answer to the right half is: {'yui_hirasawa'}

### Therefore this statement; B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$) is $\boxed{correct}$
since if you combine both halves($(B \cap A) \cup (B \cap A^{c})$), it results to: {'emu_otori','yui_hirasawa'}, which is the same as B={emu_otori',yui_hirasawa'}

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori','yui\n hirasawa'} 

drawVenn2(A,B,labels=('A', 'B'))

# (B ∩ A)
intersection_1 = B.intersection(A)

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)

# (B ∩ Aᶜ)
intersection_2 = B.intersection(complement_A)

# (B ∩ A) ∪ (B ∩ Aᶜ)
result = intersection_1.union(intersection_2)

# Comparing elements in B with the result
is_correct = result == B
print("Is B = (B ∩ A) U (B ∩ Aᶜ) correct?", is_correct)
```

A = {'emu_otori', 'hoshino_takanashi'} \

- so if $ x \in B $ are: $B$={'emu_otori','yui_hirasawa'}
- and $x \in A$ are: $A$={'emu_otori', 'hoshino_takanashi'}
- then $B \setminus A$ is: $x \in$ {'yui_hirasawa'}

So the answer to the left half is: {'yui_hirasawa'}

- if $B$ is: $B$={'emu_otori','yui_hirasawa'}
- and $x \in A^{c}$ is: $A^{c}$={'yui_hirasawa'}
- then $B \cap A^{c}$ is: {'yui_hirasawa'}

So the answer to the right half is: {'yui_hirasawa'}

- and now if you combine both halves($(B \cap A) \cup (B \cap A^{c})$), it results to: {'emu_otori','yui_hirasawa'}, which is the same as B={emu_otori',yui_hirasawa'}

### Therefore this statement; $B \setminus A $ = $B \cap$ $A^{c}$ is $\boxed{correct}$
since both halves are: {'yui_hirasawa'} = {'yui_hirasawa'} 

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

# B\A
B_minus_A = B - A

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)
        
# (B ∩ Aᶜ)
intersection = B.intersection(complement_A)

is_correct = B_minus_A == intersection
print("The equation B\\A = B ∩ Aᶜ is ", is_correct)
```

4. A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)

A = {'emu_otori', 'hoshino_takanashi'} \
B = {'emu_otori','yui_hirasawa'} 

- so if $ x \in A $ are: $A$={'emu_otori','yui_hirasawa'}
- and $x \in B$ are: $B$={'emu_otori','yui_hirasawa'}
- then $B \cup A$ is: $x \in$ {'emu_otori', 'hoshino_takanashi', 'yui_hirasawa'}

So the answer to the left half is: {'emu_otori', 'hoshino_takanashi', 'yui_hirasawa'}

- if $B$ is: $B$={'emu_otori','yui_hirasawa'}
- and $x \in A^{c}$ is: $A^{c}$={'yui_hirasawa'}
- then $B \cap A^{c}$ is: {'yui_hirasawa'}
- then if you combine the result of $B \cap A^{c}$ to A it results to: $A$={'emu_otori', 'hoshino_takanashi','yui_hirasawa}

So the answer to the right half is: {'emu_otori', 'hoshino_takanashi','yui_hirasawa}

### Therefore this statement; $A \cup B$ = $A \cup$ ($B \cap A^{c}$) is $\boxed{correct}$
since both halves are: {'emu_otori', 'hoshino_takanashi','yui_hirasawa} = {'emu_otori', 'hoshino_takanashi','yui_hirasawa}

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

# A ∪ B
union_A_B = A.union(B)

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)

# B ∩ Aᶜ
intersection_B_A_complement = B.intersection(complement_A)

# A ∪ (B ∩ Aᶜ)
union_A_intersection_B_A_complement = A.union(intersection_B_A_complement)

is_correct = union_A_B == union_A_intersection_B_A_complement
print("The equation A U B = A U (B ∩ Aᶜ) is ", is_correct)
```

6. |A $\cup$ B| = |A| + |B| - |A $\cap$ B|

A = {'emu_otori', 'hoshino_takanashi'} \
B = {'emu_otori','yui_hirasawa'} 

- so if |$A$| is $\boxed{2}$
- and |$B$| is also $\boxed{2}$
- then |A $\cup$ B| is: $\boxed{3}$ since each has 1 similar value while having two separate values

So the answer to the left half is a cardinality of 3

- if |$B$| is: $\boxed{2}$
- while |$A$| is: $\boxed{2}$
- then |$A$| + |$B$| would be: $\boxed{4}$ (I firstly did the addition since even though both (- & +) have the same importance in the operator precedence, I followed suit the (left to right) importance from the operator precedence)
- on the other hand, the |A $\cap$ B| is 1, since their is only $\boxed{1}$ similar value in them: 'emu_otori'
- so if we subtract both the: "|$A$| + |$B$|" and the "|A $\cap$ B|" then we get 3 since: (2 + 2) - 1 = $\boxed{3}$

So the answer to the right half is a cardinality of 3

### Therefore this statement; |A $\cup$ B| = |A| + |B| - |A $\cap$ B| is $\boxed{correct}$
since both halves are 3

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

union_cardinality = len(A.union(B)) # 3
A_cardinality = len(A) # 2
B_cardinality = len(B) # 2
intersection_cardinality = len(A.intersection(B)) # 1

is_correct = (union_cardinality == A_cardinality + B_cardinality - intersection_cardinality) # 3 == (2 + 2) - 1
print("The equation is", is_correct)
```
