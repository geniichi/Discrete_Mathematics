```python
from matplotlib_venn import venn2
from matplotlib import pyplot as plt
```

```python
def drawVenn2(A,B,labels = ('A', 'B')):
    '''
    listing elements in Venn diagram of two sets
    '''
    v2=venn2([A,B],set_labels = labels)
    v2.get_label_by_id('01').set_text(', '.join(map(str,B-A)))
    v2.get_label_by_id('10').set_text(', '.join(map(str,A-B)))
```

## 1. A = {2,3,4,5} and B={ $x \in \N  | x $ is even }

The statement above is not true $( A \not\subset B)$ since: 
- if $A$ = {2,3,4,5} has an odd number: 3,5
- and $B$ = { $x \in \N  | x $ is even } meaning it holds an infinite amount of only even numbers
- then if $A$ has odd numbers, and $B$ only has even numbers, then the statement "$( A \not\subset B)$" is false since by the definition of proper subset, set $B$ should have all the elements of set $A$.

```python
A = {2,3,4,5}
B = {2,4,6,8,10,12} 

if A.issubset(B) and A != B:
    print("Set A is a proper subset of set B.")
else:
    print("Set A is not a proper subset of set B.")

drawVenn2(A, B,labels=('A', 'B'))
```

The venn diagram above tells us that set A has items that are not in set B which is just considered a subset and not a proper subset
